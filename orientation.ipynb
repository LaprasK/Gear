```python
from scipy.spatial import Voronoi

def get_neighbors(v, p, pm=None):
    """ give neighbors in voronoi tessation v of point id p
        if already calculated, pm is point mask
    """
    if pm is None:
        pm = vor.ridge_points == p
    else:
        pm = pm[p]
    pts = v.ridge_points[np.any(pm, 1)]
    return pts[pts != p]
```

```python
bins = 50
locphi = []
locphi_err = []
globphi = []
globphi_err = []
globphi_stderr = []
diff_hists = {}
diff_std = []
plt.figure(1)
plt.title('orientation differences between delaunay neighbors')
plt.xlabel('angle difference (radian)')
for n, d in sorted(alldata.iteritems()):
    frames = np.unique(d['data']['f'])
    dorient = []
    locphi_n = []
    globphi_n = []
    globphi_err_n = []
    for f in frames:
        # masking/filtering:
        masks = (d['data']['f']==f) & d['omask']
        maskn = np.count_nonzero(masks)
        pts = d['data'][masks][['x','y']].view((float,2))
        orients = d['odata']['orient'][masks]
        assert pts.shape == (maskn, 2)
        # voronoi building:
        vor = Voronoi(pts)
        pairs = vor.ridge_points
        # odiff:
        dorient.extend(corr.dtheta(*orients[pairs].T, sign=True))
        # local phi:
        phi_neigh = np.abs([np.mean(np.exp(4*orients[get_neighbors(vor, i)]*1j))
                            for i in range(maskn)])
        locphi_n.append(np.abs(np.mean(phi_neigh)))
        # global phi:
        globphi_nf, globphi_err_nf = corr.global_particle_orientational(orients)
        globphi_n.append(globphi_nf)
        #globphi_err_n.append(globphi_err_nf)
    # odiff:
    dorient = np.asarray(dorient)
    diff_std.append(dorient.std())
    diff_hists[n] = plt.hist(dorient, bins, (-tau/8, tau/8), True,
                             histtype='step', label=str(n/500.), color=cm.jet(n/500.), lw=3)[:-1]
    locphi_n = np.asarray(locphi_n)
    locphi.append(locphi_n.mean()) #local phi
    locphi_err.append(locphi_n.std())#/np.sqrt(locphi_n.size/20.))
    globphi_n = np.abs(globphi_n)
    globphi.append(globphi_n.mean()) #global phi
    globphi_err.append(globphi_n.std())#/np.sqrt(globphi_n.size/10.))
plt.legend(loc='best', ncol=2)
plt.savefig('../writing/odiff_delaunay.pdf')
```

For a global measure of orientational order across the entire disk, we can measure the order parameter for particle orientation, given by $\Phi_m = \langle e^{im\theta}\rangle$, average is over all particles in a given frame.  This leaves us with a complex value with an arbitrary phase.  To obtain the full average, we take the magnitude of $\Phi_m$ and average that over all frames for a given density.

```python
ns, phis, phierr = zip(*[chain.from_iterable(([n], corr.global_particle_orientational(d['odata'][d['omask'] & (d['data']['f']==f)]['orient'])))
                       for n, d in sorted(alldata.iteritems()) for f in np.unique(d['data']['f'])])
ns, phis, phierr = map(np.array, [ns, phis, phierr])
phis, phierr = np.array([[np.abs(phis[ns==n]).mean(),
                          np.sqrt(phierr[ns==n].mean()**2 + (phis[ns==n].std()/(ns==n).sum())**2)]
                         for n in np.unique(ns)]).T
ns = np.unique(ns)

plot(ns.real/500., phis, 'o', label='abs')
errorbar(ns.real/500., phis, phierr, fmt=None, ecolor='b')
#plot(ns.real, np.angle(phis), '.', label='ang')
legend()
title("Global particle orientation parameter")
xlabel("density")
savefig('global_phi_4.pdf')
```

To determine the extent of particle orientational order, we use the $m$-fold particle orientational correlation function, which is given by $g_m(r) = \langle \cos[m (\theta_r - \theta_0 )] \rangle$, where the average is over all pairs of particles separated by distance between $r$ and $r + dr$, where $dr$ is implemented as ``binsize`` below.  Hilbert transform is also applied

```python
ss = 92.
rmax = 10*ss # r of dish is ~12.7
dr = ss*.1
rs = np.arange(ss, rmax, dr)
g4s = {}
for n, d in sorted(alldata.iteritems()):
    fs = np.unique(d['data']['f'])
    gr = np.empty(fs.shape+rs.shape, float)
    for f in fs:
        mask = (d['data']['f']==f)&d['omask']
        positions = d['data'][mask][['x','y']].view((float,2))
        dists, od = corr.orient_corr(positions, d['odata']['orient'][mask], m=4, rmax=rmax)
        for i, r in enumerate(rs):
            rmask = np.abs(dists-r) < dr
            gr[f,i] = np.mean(od[rmask])
    gr = gr.mean(0)
    g4s[n] = (gr, rs)
```
