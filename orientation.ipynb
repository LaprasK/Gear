```python
def apply_hilbert(a, sig=None, full=False):
    assert a.ndim==1
    if sig is None:
        sig = a.size/10.
    if sig:
        a_smoothed = gaussian_filter(a, sig, mode='reflect')
    else:
        a_smoothed = 0
    
    h = hilbert(a - a_smoothed)
    if full:
        return h, a_smoothed
    else:
        return np.abs(h) + a_smoothed
```

```python
def exp_decay(t, amp=1, tau=1., asymp=0.):
    assert np.all(t >= 0)
    return amp*np.exp(-t/tau) + asymp

def alg_decay(t, amp=1, pow=1., asymp=0.):
    return amp*np.power(t, -pow) + asymp

def log_decay(t, amp=1, tau=1., asymp=0.):
    return asymp - amp*np.log(t/tau)

decays = {'exp':exp_decay, 'alg':alg_decay, 'log':log_decay}
```

```python
x = np.linspace(0,1,10001)[1:]
plot(x, exp_decay(x), label='exp')
plot(x, alg_decay(x), label='alg')
plot(x, log_decay(x), label='log')
legend()
xlim(0,1); ylim(0,10)
```
