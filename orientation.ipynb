To determine the extent of translational order, we measure the pair correlation function $g(r)$.  This is implemented as a weighted histogram of the distances between all pairs of particles.  The amplitude envelope of $g(r)$ gives a decay shape and length scale, which we can use to quantify the range of the ordering.  To obtain the envelope, we apply a Hilbert transform to the curve after removing 

```python
for n, d in sorted(alldata.iteritems())[-1:]:
    dr = 1/10.
    gs, rgs = corr.build_gs(d['data'], None, 1, dr=dr)
    g, dg, rg = corr.avg_hists(gs, rgs)
    g = g[rg>92]
    rg = rg[rg>92]
    # convert pixels to squares:
    r = rg/92.
    x = r - 1

    #gm = np.convolve(g, np.ones(2/dr)*dr/2., 'same')
    h = corr.apply_hilbert(g, 1/dr, #full=True
                      )
    step(r, g, ls='-', where='pre', label=str(n/500.), color=cm.jet(n/500.), lw=2)
    plot(r, h, color=cm.jet(n/500.), lw=2,)# label='hilbert')
    #plot(r, g, '.')
    #plot(r, gm, label='moving avg')
    #plot(r, gm+h.real, label='real')
    #plot(r, gm+h.imag, label='imag')
    #plot(r, h*h.conj(), label='sqr'
    #errorbar(r, g, color=cm.jet(n/500.))
    print n/500.
    w = np.sqrt(h*r)
    plot(r, w*.1, ':', label='weight')
    for name, decay in decays.iteritems():
        p, cov = curve_fit(decay, x, h, (1., 1., 0), sigma=w)
        plot(r, decay(x, *p), label=name)
        resid = (h - decay(x, *p))
        chi2 = np.dot(resid, resid)
        print '\t', name, "length: {:.3f}, amplitude: {:.3f}, offset: {:.3f}".format(*p),\
              ", chi^2: {:.3e}".format(chi2)

#A, A0 = .025, .003
#T, T0 = 2., 1.
#W = 2*np.pi * .000000001
#plot(rg/92., A0 + A*np.exp((T0-r)/T)*np.cos(W*r))
legend()
title("Pair correlation function $g(r)$")
xlabel("$r/s$")
#xlim(0.9, 1.1)
```

To determine the extent of particle orientational order, we use the $m$-fold particle orientational correlation function, which is given by $g_m(r) = \langle \cos[m (\theta_r - \theta_0 )] \rangle$, where the average is over all pairs of particles separated by distance between $r$ and $r + dr$, where $dr$ is implemented as ``binsize`` below.  Hilbert transform is also applied

```python
for n, d in sorted(alldata.iteritems()):
    positions = corr.get_positions(d['data'][d['omask']], 1)
    d, od = corr.orient_corr(positions, d['odata']['orient'][(d['data']['f']==1)&d['omask']])
    #scatter(d/92., od, marker='.')
    si = d.argsort()
    binsize = 92.  #92 = particle size
    od_smooth = gaussian_filter(od[si], binsize)
    plot(d[si]/92., od_smooth, label=str(n/500.), color=cm.jet(n/500.))
    plot(d[si]/92., corr.apply_hilbert(od_smooth, binsize*4), '--', color=cm.jet(n/500.))
    title(r'Particle orientation correlation $g_4(r)$')
    xlabel('$r/s$')
    legend()
```

For a global measure of orientational order across the entire disk, we can measure the order parameter for particle orientation, given by $\Phi_m = \langle e^{im\theta}\rangle$, average is over all particles in a given frame.  This leaves us with a complex value with an arbitrary phase.  To obtain the full average, we take the magnitude of $\Phi_m$ and average that over all frames for a given density.

```python
ns, phis, phierr = zip(*[chain.from_iterable(([n], corr.global_particle_orientational(d['odata'][d['omask'] & (d['data']['f']==f)]['orient'])))
                       for n, d in alldata.iteritems() for f in np.unique(d['data']['f'])])
ns, phis, phierr = map(np.array, [ns, phis, phierr])
phis, phierr = np.array([[np.abs(phis[ns==n]).mean(),
                          np.sqrt(phierr[ns==n].mean()**2 + (phis[ns==n].std()/(ns==n).sum())**2)]
                         for n in np.unique(ns)]).T
ns = np.unique(ns)
```
