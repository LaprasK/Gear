For a global measure of orientational order across the entire disk, we can measure the order parameter for particle orientation, given by $\Phi_m = \langle e^{im\theta}\rangle$, average is over all particles in a given frame.  This leaves us with a complex value with an arbitrary phase.  To obtain the full average, we take the magnitude of $\Phi_m$ and average that over all frames for a given density.

To determine the extent of translational order, we measure the pair correlation function $g(r)$.  This is implemented as a weighted histogram of the distances between all pairs of particles.  The amplitude envelope of $g(r)$ gives a decay shape and length scale, which we can use to quantify the range of the ordering.  To obtain the envelope, we apply a Hilbert transform to the curve after removing

To determine the extent of particle orientational order, we use the $m$-fold particle orientational correlation function, which is given by $g_m(r) = \langle \cos[m (\theta_r - \theta_0 )] \rangle$, where the average is over all pairs of particles separated by distance between $r$ and $r + dr$, where $dr$ is implemented as ``binsize`` below.  Hilbert transform is also applied

```python
rmax = 10*S # r of dish is ~12.7
dr = .5*S
rbins = np.linspace(S-dr/2, rmax+dr/2, np.round((rmax-S+dr)/dr))
rs = .5*(rbins[1:] + rbins[:-1])
m = 4
gms[m] = {}
for n, d in sorted(alldata.iteritems()):
    fs = np.unique(d['data']['f'])
    gmr = np.empty(fs.shape+rs.shape, float)
    for f in fs:
        mask = (d['data']['f']==f) & d['omask']
        positions = np.column_stack([d['data'][mask]['x'], d['data'][mask]['y']])
        dists, od = corr.orient_corr(positions, d['odata']['orient'][mask], m=m, rmax=rmax)
        nr, _ = np.histogram(dists, rbins)
        gmr[f] = np.histogram(dists, rbins, weights=od)[0]/nr
    gms[m][n] = (gmr.mean(0), rs)
```

```python
figsize(5,4)
m = 4
try:
    gmlengths[m] = []
except NameError:
    gmlengths = {}
    gmlengths[m] = []
for n, (gmr, rs) in sorted(gms[m].items()):
    figure(1)
    plot(rs/S, gmr, label=format(n/N, '.2%'), color=cm.jet(n/N), lw=2)
    figure(2)
    semilogy(rs/S, np.where(gmr > 0, gmr, np.nan), label=format(n/N, '.2%'), color=cm.jet(n/N))
    
    p, cov = curve_fit(decays['exp'], rs/S - 1, gmr, [1])
    print "N = {}, decay length = {}, cov = {}".format(n, p[0], cov)
    if np.isfinite(cov):
        gmlengths[m].append((n/N, p[0]))
        figure(1)
        plot(rs/S, decays['exp'](rs/S - 1, *p), color=cm.jet(n/N))
        figure(2)
        semilogy(rs/S, decays['exp'](rs/S - 1, *p), color=cm.jet(n/N))
        figure(3)
        plot(rs/S, (gmr - decays['exp'](rs/S - 1, *p))/decays['exp'](rs/S - 1, *p),
             label=format(n/N, '.2%'), color=cm.jet(n/N), lw=2)

figure(1)
title(r'Particle orientation correlation $g_{}(r)$'.format(m))
xlabel('$r/s$')
xlim(0, rmax/S)
ylim(-.25, 1)
#legend(loc='best', ncol=2, fontsize='small')
savefig(savedir+"g{}_r.pdf".format(m))

figure(2)
title(r'Particle orientation correlation $g_{}(r)$'.format(m))
xlabel('$r/s$')
ylim(1e-4, 1)
xlim(1, rmax/S)
#legend(loc='best', ncol=2, fontsize='small')
savefig(savedir+"g{}_r_semilog.pdf".format(m))

plt.figure(3)
title('residuals')
ylabel("$g_{}(r) - $exp fit".format(m))
xlabel('$r/s$')
ylim(-5,5)
savefig(savedir+"g{}_r_resid.pdf".format(m))

plt.figure(4)
plt.plot(*zip(*gmlengths[m]), marker='o')
plt.xlabel('density')
plt.ylabel("$g_{}(r)$ length scale".format(m))
plt.title("fits for $g_{}(r)$".format(m))
plt.savefig(savedir+"g{}_fits.pdf".format(m))
```
