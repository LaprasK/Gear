```python
from PIL import Image as im
from scipy.ndimage.filters import convolve, correlate
from skimage.filter import canny
from skimage.morphology import disk
import numpy as np
from matplotlib import pyplot as plt
```

```python
asize = 50
dsize = asize/5
a = np.zeros((asize,asize))
a[3:3+2*dsize+1,10:10+2*dsize+1] = disk(dsize)
dsize /= 2
a[asize-3-2*dsize-1:asize-3,asize-10-2*dsize-1:asize-10] = disk(dsize)
a += np.array(np.random.random(a.shape) < .5, dtype=np.int)
a -= np.array(np.random.random(a.shape) < .5, dtype=np.int)
```

```python
rod = np.array([[1,1,1,1,1]])
d = disk(dsize)
c = convolve(a,d, mode='constant')
e = canny(a)
ef = canny(c)
```

```python
plt.imshow(a,interpolation='nearest')
plt.colorbar()
```

```python
plt.imshow(c,interpolation='none')
plt.colorbar()
```

```python
plt.imshow(ef,interpolation='none')
```
